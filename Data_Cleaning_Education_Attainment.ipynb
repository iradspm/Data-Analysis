```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Education_attainment.csv",na_values="---")
data
```

```python
pd.isna(data).sum()
```

```python
#replace nan values with mean of each column
data1=data.fillna(data.mean())
data1
```

```python
pd.isna(data1).sum()
```

```python
#check for duplicates
data1.duplicated().sum()
```

```python
#filter columns
columns=['Sex','White','Black','Hispanic','Asian','Pacific Islander']
data2=data1[columns]
data2
```

```python
#remove sex 'A' representing both
data3=data2[(data2['Sex'] !='A')]
data3
```

```python
#using classification model, Sex becomes our class(target) value, 
#so M gender will be represented by 1 and F gender represented by M
data3.dtypes
```

```python
data3['Gender']= np.where(data3["Sex"].str.contains("M"), 1, 0)
data3['Gender']
```

```python
columns=['White','Black','Hispanic','Asian','Pacific Islander','Gender']
data4=data3[columns]
```

```python
data4
```

```python
#split data into inputs and output
s_data=data4.values
x=s_data[:,0:5]
y=s_data[:,5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.33, random_state=2)
print(x_train.shape,y_train.shape) #get data used for training
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
#print(knn.predict(x_test))
print(knn.score(x_test,y_test))
```
