```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': [1, 2, 3, 4],
'B': [10, 20, 30,40],
'C': [20,40, 80,90]}, index=['Row 1','Row 2','Row 3','Row 4'])
df
```

```python
#using lambda syntax: df.apply(lambda x)--sum rows
df['D']=df.apply(lambda x:x.sum(),axis=1)
df
```

```python
#sum columns
df.loc['Total']=df.apply(lambda x:x.sum(),axis=0)
df
```

```python
#multiply by 2 column c using lambda
df['E']=df['C'].apply(lambda x:x*2)
df
```

```python
#applymap
df.applymap(lambda x: x**2)
df
```
