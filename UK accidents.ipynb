## Analysis/Visualization of the dataset

```python
#Get unique features of textual columns
pedestrians = data_1['Pedestrian_Crossing-Human_Control'].value_counts()
pedestrians
```

```python
plt.figure(figsize =(15,7))
plt.title("Pedestrian crossing human control analysis")
ax=sns.countplot(x='Pedestrian_Crossing-Human_Control', data=data_1, order=pedestrians.index)
ax.set_xticklabels(['Within 50 mtrs','authorized_persons','school_crossing'])
plt.show()
```

Most accidents occur within 50 metres when pedestrians are crossing

```python
pedestrians_1=data_1['Pedestrian_Crossing-Physical_Facilities'].value_counts()
pedestrians_1
```

```python
plt.figure(figsize =(15,7))
plt.title("Pedestrian Crossing Physical Facilities")
ax=sns.countplot(x='Pedestrian_Crossing-Physical_Facilities', data=data_1, order=pedestrians_1.index)
ax.set_xticklabels(['Within 50 mtrs','traffic signal junction','non junction','zebra crossing',
                    'central refuge','Footbridge or subway'])
plt.show()
```

```python
light=data_1['Light_Conditions'].value_counts()
light
```

```python
plt.figure(figsize =(15,7))
plt.title("Light conditions")
ax=sns.countplot(x='Light_Conditions', data=data_1, order=light.index)
ax.set_xticklabels(['DayLight-Light present','Darkness-light present','Darkness-No light','Darkness-unk light',
                    'Darkness-light present but unlit'])
plt.show()
```

Most of accidents occur during the day when the light is present.

Accidents occur mostly when the weather is fine without high winds.

```python
'''Scatter plot! We expect to see on average, more vehicles = more casualties (the boxes) 
all the dots are outliers''' 
plt.figure(figsize =(15,7))
data_1.plot.scatter(x="Number_of_Vehicles", y="Number_of_Casualties")
plt.show()
```

```python
# check speed limit
import numpy as np
plt.figure(figsize =(15,7))
data_1.groupby('Speed_limit')['Number_of_Casualties'].mean().plot(kind='bar',grid=False)
plt.xticks(np.arange(8), 
           ['10mph', '15mph', '20mph', '30mph', '40mph', '50mph', '60mph', '70mph'], 
           rotation='horizontal')
plt.ylim((0.6,2.0))
plt.xlabel(''), plt.ylabel('Average Number of Casualties\n')
plt.title('\nAverage Number of Casualties by Speed Limit\n', fontweight='bold')
sns.despine(top=True, right=True, left=True, bottom=True);
```

It can be noted that most number of casualties occur when driving at speed of 70 mph and occur lowest when driving at 15 mph.

# Conclusion

Most accidents in UK occurs when:-
1. None of pedestrians are crossing human control within 50 metres.
2. No pedestrians are crossing physical facilities
3. There is daylight street light present
4. When the weather fine without high winds
6. Majority of accidents involved police officers attending areas of the scene
7. Accidents occurs of day 6, at 1700 hours
8. Most accidents occurs within November.
