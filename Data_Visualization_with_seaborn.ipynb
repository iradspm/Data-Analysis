```python
#imports
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load Education attainment dataset and replace '---' with NaN
df=pd.read_csv("Education_attainment.csv",na_values="---") 
```

```python
df
```

```python
#get NaN values and replace them with 0
df.notnull()
```

```python
df.notnull().sum()
```

```python
#fill null values with 0
df1=df.fillna(0)
df1
```

```python
pd.isna(df1) #check for null values
```

```python
#count null values in df1 for each column
pd.isna(df1).sum()
```

```python
df1.head()
```

```python
#visualize 
#sex and Total
columns=['Sex','Total']
df2=df1[columns]
df2
```

```python
#draw bar chart and provide title to the chart
plt.title("Bar chart of Sex vs Total over the years")
# plt.ylabel("Percentage")
# plt.xlabel("Sex")
sns.barplot(x='Sex',y='Total', data=df2)
```

```python
#bar chart of min degree and total-detecting minimum degree mostly awarded
columns=['Min degree','Total']
df3=df1[columns]
sns.barplot(x='Min degree', y='Total', data=df3)
```

```python
#visualize the trend of each degree being awarded over the years
columns=['Min degree', 'Year','Total']
df4=df1[columns]
df4
```

```python
#multiple line plots
high_school=df4[(df4['Min degree']=='high school')] #select high school degree
high_school.head()
```

```python
#select bachelors
bachelors=df4[(df4['Min degree']=="bachelor's")]
bachelors.head()
```

```python
#select associates and masters
associates=df4[(df4['Min degree']=="associate's")]
masters=df4[(df4['Min degree']=="master's")]
```

```python
#plot line for high school, bachelors, associates and masters
sns.lineplot(x='Year',y='Total', data=high_school)
sns.lineplot(x='Year',y='Total', data=bachelors)
sns.lineplot(x='Year',y='Total', data=associates)
sns.lineplot(x='Year',y='Total', data=masters)
```

```python
sns.lineplot(x='Year', y='Total', hue='Min degree', data=df1) #simpler method for plotting multiple lines with a key
```

```python
#visualizing min degree from 2000 to 2020
#hue attribue specifies column providing 'key' of the line plotted
sns.lineplot(x='Year',y='Total', data=df1[(df1['Year']>=2000) & (df1['Year']<=2020)],hue='Min degree') 
```
