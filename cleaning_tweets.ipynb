```python
import pandas as pd
import numpy as np
```

```python
#read csv file
file=pd.read_csv("data.csv",encoding="latin-1")
file.head()
```

```python
#filter text column alone
columns=['text']
df=file[columns]
```

```python
df.head()
```

```python
'''
remove all characters in the tweet except the ASCII
ASCII contains the numbers from 0-9, the upper and lower case English letters from A to Z, 
and some special characters.
'''
```

```python
import re
def replace_foreign_characters(s):
    return re.sub(r'[^\x00-\x7f]',r'', s)

df['text'] = df['text'].apply(lambda x: replace_foreign_characters(x))
```

```python
df
```

```python
#remove @
df['text'] = df['text'].replace({'@':''}, regex=True)
df
```

```python
#remove RT
df['text'] = df['text'].replace({'RT':''}, regex=True)
df
```

```python
df
```

```python
df["text"] = df["text"].str.replace(r'^.*:s*', '') #remove usernames in the text--i.e username followed by ':'
df
```

```python
#remove duplicates
df1=df.drop_duplicates()
df1
```
